```python
import os 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
path_train = "../data/new_data"

os.listdir(path_train)
```

```python
list_files = os.listdir(os.path.join(path_train, "1290575559719166237"))
list_files
```

```python
path_upd = os.path.join(path_train, "1290575559719166237")
```

```python
img = np.load(os.path.join(path_upd, 'img.npy'))
human_pixel_mask = np.load(os.path.join(path_upd, 'human_pixel_mask.npy'))
human_individual_mask = np.load(os.path.join(path_upd, 'human_individual_mask.npy'))
```

```python
plt.figure(figsize=(18, 6))
ax = plt.subplot(1, 4, 1)
ax.imshow(img)
ax.set_title('False color image')

ax = plt.subplot(1, 4, 2)
ax.imshow(np.mean(human_individual_mask, axis=3), interpolation='none')
ax.set_title('Concat individual_mask')

ax = plt.subplot(1, 4, 3)
ax.imshow(img)
ax.imshow(human_pixel_mask, interpolation='none')
ax.set_title('Results pixel mask');

ax = plt.subplot(1, 4, 4)
ax.imshow(img)
ax.imshow(human_pixel_mask, cmap='Reds', alpha=.4, interpolation='none')
ax.set_title('Contrail mask on false color image');
```
